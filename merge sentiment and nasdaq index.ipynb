```python
import pandas as pd
```

```python
sentiment_output = pd.read_excel('/Users/ngocnguyen/Downloads/output_2018.xlsx')
nasdaq_index = pd.read_csv('/Users/ngocnguyen/Desktop/CPP/Capstone/nasdaq_index.csv')
```

```python
# Merge the datasets on "Date"
merged_df = sentiment_output.merge(nasdaq_index[["Date", "Index Price"]], on="Date", how="left")

# Reorder columns: Move "Index Price" right after "Date"
cols = ["Date", "Index Price"] + [col for col in merged_df.columns if col not in ["Date", "Index Price"]]
merged_df = merged_df[cols]

# Display first few rows
merged_df.head(10)
```

```python
merged_df = merged_df.dropna(subset=["Index Price"])
merged_df.head()
```
